```python
import pandas as pd
import joblib
```

Predicting next year's points using this year's stats

```python
# Load the trained model
model = joblib.load('random_forest_model.pkl')
```

```python
df = pd.read_csv(r'Data/seasonstats_imputed.csv')
```

```python
# Bring in data from the 2023/2024 season
current_season_df = df[df['Season'] == '2023/2024']
```

```python
current_season_df.shape
```

```python
# Ensure the DataFrame has the same structure as the training data
X_current_season = current_season_df.drop(columns=['Unnamed: 0', 'Season', 'Squad', 'Pts'])
```

```python
X_current_season.shape
```

```python
# Make predictions for the next season's points
predicted_points = model.predict(X_current_season)
```

```python
current_season_df
```

```python
current_season_df['Predicted Points 2024/2025'] = predicted_points
```

```python
current_season_df.sort_values(by='Predicted Points 2024/2025', ascending=False)
```
